# Plotting und Sigmoid Funktion

Wir lernen heute Funktionen mit `matplotlib` und `seaborn` zu plotten. Als Beispiel betrachten wir die **Sigmoid Funktion**, die wesentlicher Bestandteil der **Logistic Regression** ist.

## Sigmoid Funktion

Die 1-dimensionale Sigmoid Funktion ist wie folgt definiert:

$$\begin{align}
\mathbb{R} &\rightarrow \mathbb{R}\\
x &\mapsto \frac{1}{1 + e^{-x}}
\end{align}$$

Um zu verstehen, wie sich diese Funktion verhält betrachtem wir die einzelnen Terme.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
plt.ioff()
```

Jetzt definieren wir die einzelnen Funktionen.

```python
def neg_exp(x):
    return math.exp(-x)
```

```python
def neg_exp_one(x):
    return 1 + math.exp(-x)
```

```python
def sigmoid(x):
    result = neg_exp_one(x)
    result = 1 / result
    return result
```

Vorbereiten der arrays für den Plot.

```python
x = np.linspace(-2, 2, 200)
vfunc_1 = np.vectorize(neg_exp)
y_1 = vfunc_1(x)
vfunc_2 = np.vectorize(neg_exp_one)
y_2 = vfunc_2(x)
vfunc_3 = np.vectorize(sigmoid)
y_3 = vfunc_3(x)
```

## matplotlib

Wir erstellen einen `subplot`, der die drei Funktionen gleichzeitig darstellt.

```python
fig, ax = plt.subplots()
ax.plot(x,y_1,x,y_2,x,y_3)
plt.show()
```

Nun wissen wir wie Sigmoid zustande kommt und betrachten die Funktion alleine. Dafür wählen wir einen gößeren Definitionsbereich, um das Verhalten sichtbar zu machen.

```python
x = np.linspace(-10, 10, 200)
vfunc = np.vectorize(sigmoid)
y = vfunc(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

## 2-dim Sigmoid und 3-Plot

Die Sigmoid Funktion kann auch auf höhere Dimemsionen verallgemeinert werden. Hier die 2-dim Version

$$\begin{align}
\mathbb{R}^2 &\rightarrow \mathbb{R}\\
\left(
\begin{array}{c}
x_1\\
x_2\\
\end{array}
\right)
 &\mapsto \frac{1}{1 + e^{-(x_1 + x_2)}}
\end{align}$$

```python
import seaborn as sn
```

```python
# Define the sigmoid function
def sigmoid(x, y):
    return 1 / (1 + np.exp(-(x + y)))
```

```python
#sn.set_style('whitegrid')
X_VAL = np.linspace(-5, 5, 100)
Y_VAL = np.linspace(-5, 5, 100)
```

```python
# Create a grid
X, Y = np.meshgrid(X_VAL, Y_VAL)
```

```python
# Compute the sigmoid function for each pair of x and y coordinates
Z = sigmoid(X, Y)
```

```python
axes = plt.axes(projection='3d')
axes.plot_surface(X, Y, Z,cmap='viridis')
# Set labels and title
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Sigmoid Value')
axes.set_title('3D Sigmoid Function')
# show
plt.show()
```
