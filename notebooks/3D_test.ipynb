```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.ioff()
```

```python
# Define the sigmoid function
def sigmoid(x, y, alpha):
    return 1 / (1 + np.exp(-alpha*(x + y)))
```

```python
# Create a grid of x and y values
x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
```

```python
alpha = 0.1
```

```python
# Compute the sigmoid function for each pair of x and y coordinates
Z = sigmoid(X, Y, alpha)
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
```

```python
# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
```

```python
# Add a colorbar which maps values to colors
fig.colorbar(surf, ax=ax, label='Sigmoid Value')
```

```python
# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Sigmoid Value')
ax.set_title('3D Sigmoid Function')
```

```python
# Show the plot
plt.show()
```
